```python
# =====================================
# SKIN CONCERNS MULTI-LABEL CLASSIFIER
# =====================================

# 1️⃣ Install / import dependencies
!pip install tensorflow pandas numpy scikit-learn matplotlib --quiet

import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# 2️⃣ Set dataset folder path
DATASET_DIR = "/content/skin_concern"  # Change if you mount Drive
IMG_SIZE = (224, 224)
BATCH_SIZE = 16

# 3️⃣ Create labels CSV from folder structure
# Folder structure: skin_concern/<concern>/<images>
concerns = os.listdir(DATASET_DIR)
filepaths = []
labels_list = []

for idx, concern in enumerate(concerns):
    concern_folder = os.path.join(DATASET_DIR, concern)
    if os.path.isdir(concern_folder):
        for img_file in os.listdir(concern_folder):
            filepaths.append(os.path.join(concern_folder, img_file))
            labels = [0]*len(concerns)
            labels[idx] = 1
            labels_list.append(labels)

# Create DataFrame
df = pd.DataFrame(labels_list, columns=concerns)
df['filepath'] = filepaths

# Shuffle dataset
df = df.sample(frac=1).reset_index(drop=True)
df.head()

# 4️⃣ Split into train & validation
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# 5️⃣ Data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

# 6️⃣ Generators
def create_generator(datagen, dataframe):
    return datagen.flow_from_dataframe(
        dataframe,
        x_col='filepath',
        y_col=concerns,
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='raw',  # raw = multi-label
        shuffle=True
    )

train_gen = create_generator(train_datagen, train_df)
val_gen = create_generator(val_datagen, val_df)

# 7️⃣ Build model (Transfer Learning with EfficientNetB0)
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_model.trainable = False  # Freeze base for initial training

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.3)(x)
output = Dense(len(concerns), activation='sigmoid')(x)  # Sigmoid for multi-label

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer=Adam(learning_rate=1e-3),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()

# 8️⃣ Train model
EPOCHS = 10
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS
)

# 9️⃣ Save model
MODEL_PATH = "/content/skin_concerns_model.h5"
model.save(MODEL_PATH)
print(f"Model saved to {MODEL_PATH}")

# 1️⃣0️⃣ Plot training curves
plt.figure(figsize=(10,5))
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend()
plt.show()
```
